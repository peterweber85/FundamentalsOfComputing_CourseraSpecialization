```python
import urllib2
import matplotlib.pyplot as plt
import numpy as np
import random

import helper_module as hm

%matplotlib inline

CITATION_URL = "http://storage.googleapis.com/codeskulptor-alg/alg_phys-cite.txt"



def plot_in_degree_distribution(keys, values, filename_out, color = "blue", xscale = 'log', yscale = 'log', x_low = 1e-1, x_up = 1e4, y_low = 1e-5, y_up = 1):
    
    norm = np.sum(keys)
    plt.scatter(keys, map(float,values)/norm)

    plt.title("In degree distribution of DPA, n = 28000, m = 12")
    plt.xlabel("in degree")
    plt.ylabel("Probability")

    if xscale == 'log':
        plt.xscale('log')
    
    if yscale == 'log':
        plt.yscale('log')
    
    plt.xlim(x_low, x_up)
    plt.ylim(y_low,y_up)
    plt.savefig(filename_out)
```

```python
citation_graph = hm.load_graph(CITATION_URL)
idd_citation_graph = hm.in_degree_distribution(citation_graph) 
plot_in_degree_distribution(idd_citation_graph.keys(),idd_citation_graph.values(),"idd_of_citation_graphs.png")
```

```python
n = 1000
p = 0.5
ER_graph1 = hm.make_ER_graph(n, p)
idd_ER_graph1 = hm.in_degree_distribution(ER_graph1) 
plot_in_degree_distribution(idd_ER_graph1.keys(),idd_ER_graph1.values(),"ER_graph_distribution.png", x_low = 1, x_up = n)
```

```python
def make_DPA(n,m):
    trials = n-m 
    DPA_graph = hm.make_complete_graph(m) 
        
    experiment = hm.DPATrial(m)
    for trial in range(trials):
        new_edges = experiment.run_trial(m)
        add_node = m + trial 
        DPA_graph[add_node] = new_edges
        
    return DPA_graph
```

```python
DPA = make_DPA(28000, 12)
```

```python
idd_DPA = hm.in_degree_distribution(DPA) 
plot_in_degree_distribution(idd_DPA.keys(),idd_DPA.values(), "in_degree_distribution_of_DPA", y_low = 1e-8)
```
