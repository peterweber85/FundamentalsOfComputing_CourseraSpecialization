```python
m = 5
```

```python
iterations = range(10, 1000, 10)
```

```python
from timeit import default_timer as timer

start_fast = timer()   
time_array_fast = []

for n in iterations:

    temp_UPA = hp.make_UPA(n, m)
    temp_order = hp.fast_targeted_order(temp_UPA)
    #resilience_fast = hp.compute_resilience(temp_UPA, temp_order)
    end_fast = timer()
    time_array_fast.append(end_fast - start_fast)

print "total_time: ", time_array_fast[-1]   
```

```python
from timeit import default_timer as timer

start = timer()   
time_array = []

for n in iterations:

    temp_UPA = hp.make_UPA(n, m)
    temp_order = hp.targeted_order(temp_UPA)
    #resilience_fast = hp.compute_resilience(temp_UPA, temp_order)
    end = timer()
    time_array.append(end-start)

print "total_time: ", time_array[-1] 
```

```python
f, ax1 = plt.subplots()
ax1.plot(iterations, time_array_fast, label = "fast targeted order")
ax1.plot(iterations, time_array, label = "targeted order")
ax1.legend(loc = 2)

ax1.set_xlabel("Number of nodes in the UPA graph")
ax1.set_ylabel("Running time (s)")
ax1.set_title("Running times in IPython notebook")

plt.savefig("Question3.png", dpi = 600)
```
